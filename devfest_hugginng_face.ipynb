```python
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification

raw_datasets = load_dataset("imdb")
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
model=AutoModelForSequenceClassification.from_pretrained("bert-base-cased",num_labels=2)
```

```python
sentences=['Yay I am attending devfest ', "Wellcome to the devfest noida"]
inputs=tokenizer(sentences,padding="max_length",truncation=True)
inputs
```

```python
def tokenize_function( examples):
  return tokenizer(examples["text"],padding="max_length",truncation=True)
tokenized_datasets=raw_datasets.map(tokenize_function,batched=True)
#tokenized_datasets
```

```python
small_train_dataset= tokenized_datasets["train"].shuffle(seed=42).select(range(10))
small_eval_dataset= tokenized_datasets["test"].shuffle(seed=42).select(range(10))
print(len(small_train_dataset),len(small_eval_dataset))
```

```python
from transformers import TrainingArguments
training_args=TrainingArguments("test-trainer",evaluation_strategy="epoch",report_to="none")
```

```python
from transformers import Trainer
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset
)
```

```python
trainer.train()
```

```python
import numpy as np
from datasets import load_metric
metric=load_metric("accuracy")
def compute_metrics(eval_pred):
  logits,labels=eval_pred
  predictions=np.argmax(logits,axis=-1)
  return metric.compute(predictions=predictions,references=labels)
```

```python
trainer=Trainer(
    model=model,
    args=training_args,
    train_dataset=small_train_dataset,
    eval_dataset=small_eval_dataset,
    compute_metrics=compute_metrics
)
```
